```python
# import the necessary packages
import os
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import cv2
import numpy as np
import math
from skimage import io, color
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg19 import preprocess_input
from tensorflow.keras import applications
from tensorflow.keras.models import Model
from tensorflow.python.framework import ops
import tensorflow as tf
import random

tf.enable_eager_execution()

VGG19 = applications.vgg19.VGG19(weights='imagenet', include_top=False, pooling='avg')
VGG19.trainable = False
```

```python
def list_files(folder):
    r = []
    x=0
    for root, dirs, files in os.walk(folder):
        for name in files:
            if (('.jpg' in name) or ('.png' in name)):
                r.append(os.path.join(root,name))
                x = x + 1
    return r

def tensor_to_array(tensor1):
    return tensor1.numpy()

def gram_matrix(input_tensor):
    result = tf.linalg.einsum('bijc,bijd->bcd', input_tensor, input_tensor)
    input_shape = tf.shape(input_tensor)
    num_locations = tf.cast(input_shape[1]*input_shape[2], tf.float32)
    return result/(num_locations)


def vgg_style_layers():
    style_layers = ['block1_conv1',
                'block2_conv1',
                'block3_conv1']
    outputs = [VGG19.get_layer(name).output for name in style_layers]
    model = tf.keras.Model([VGG19.input], outputs)
    return model

def centroid_histogram(clt):
    # grab the number of different clusters and create a histogram
    # based on the number of pixels assigned to each cluster
    numLabels = np.arange(0, len(np.unique(clt.labels_)) + 1)
    (hist, _) = np.histogram(clt.labels_, bins = numLabels)
    # normalize the histogram, such that it sums to one
    hist = hist.astype("float")
    hist /= hist.sum()
    # return the histogram
    return hist

def plot_colors(hist, centroids):
    # initialize the bar chart representing the relative frequency
    # of each of the colors
    bar = np.zeros((50, 300, 3), dtype = "uint8")
    startX = 0
    # loop over the percentage of each cluster and the color of
    # each cluster
    for (percent, color) in zip(hist, centroids):
        # plot the relative percentage of each cluster
        endX = startX + (percent * 300)
        cv2.rectangle(bar, (int(startX), 0), (int(endX), 50),
        color, -1)
        startX = endX

    # return the bar chart
    return bar

def get_palette(Image, Clusters):
    image = cv2.imread(Image)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    scale_percent = 25 # percent of original size
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)
    resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)

    # show our image
    plt.figure()
    plt.axis("off")
    plt.imshow(image)

    im = resized.reshape((resized.shape[0] * resized.shape[1], 3))
    clt = KMeans(n_clusters = Clusters)
    clt.fit(im)
    hist = centroid_histogram(clt)

    order = np.argsort(hist)
    new_hist, new_cols = [], []
    for i in order:
        new_hist.append(hist[i])
        new_cols.append(clt.cluster_centers_[i].tolist())
    new_hist.reverse()
    new_cols.reverse()

    bar = plot_colors(new_hist, new_cols)
    # show our color bart
    plt.figure()
    plt.axis("off")
    plt.imshow(bar)
    plt.show()
    return new_hist, new_cols

def get_colour_vect_dict(new_hist, new_cols):
    Ls,As,Bs,Covers,DiffLs,DiffAs,DiffBs,DiffCovers = [],[],[],[],[],[],[],[]
    for i in range(len(new_cols)):
        col = new_cols[i]
        rgb = [value / 255 for value in col]
        lab = color.rgb2lab(rgb)
        Ls.append(min(1,lab[0]/100))
        As.append(max(-1,min(1,lab[1]/100)))
        Bs.append(max(-1,min(1,lab[2]/100)))
        Covers.append(new_hist[i])
        if i ==0:
            base_col = new_cols[0]
            base_rgb = [value / 255 for value in base_col]
            base_lab = color.rgb2lab(base_rgb)
        if i > 0:
            DiffLs.append(min(1,lab[0]/100) - min(1,base_lab[0]/100))
            DiffAs.append(max(-1,min(1,lab[1]/100)) - max(-1,min(1,base_lab[1]/100)))
            DiffBs.append(max(-1,min(1,lab[2]/100)) - max(-1,min(1,base_lab[2]/100)))
            DiffCovers.append(new_hist[i]-new_hist[0])
        print(lab)
    vect_dict = {
        "Absolute_Luminosity":Ls+Covers,
        "Relative_Luminosity":DiffLs + DiffCovers,
        "Absolute_Colours":As+Bs+Covers,
        "Relative_Colours":DiffAs+DiffBs+DiffCovers,
    }
    return vect_dict

def create_colour_vect(vect_dict, weights):
    vec = []
    for key in vect_dict:
        weight = weights[key]
        weighted_vec = [value * weight for value in vect_dict[key]]
        vec = vec + weighted_vec
    return vec

def get_colour_dist_between_two_images(image1, image2, no_cols, colour_weights):
    hist, cols = get_palette(image1, no_cols)
    vec_dict = get_colour_vect_dict(hist, cols)
    weighted_col_vec = create_colour_vect(vec_dict, colour_weights)
    a = np.array(weighted_col_vec)
    
    hist, cols = get_palette(image2, no_cols)
    vec_dict = get_colour_vect_dict(hist, cols)
    weighted_col_vec = create_colour_vect(vec_dict, colour_weights)
    b = np.array(weighted_col_vec)
    
    return np.linalg.norm(a-b)

def get_colour_vector(image_path):
    im1 = cv2.imread(image_path)
    lab = cv2.cvtColor(im1, cv2.COLOR_BGR2LAB) #cv2.COLOR_RGB2LAB)
    hist1 = cv2.calcHist(lab, [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    return cv2.normalize(hist1, hist1).flatten()

def get_colour_corr(image1, image2):
    hist1 = get_colour_vector(image1)
    hist2 = get_colour_vector(image2)
    corr = cv2.compareHist(hist1, hist2, cv2.HISTCMP_CORREL)
    return corr

def get_content_vector(image_path):
    img = image.load_img(image_path, target_size=(512, 512))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    return VGG19.predict(x)[0]

def get_content_corr(image1,image2):
    vect1 = get_content_vector(image1)
    vect2 = get_content_vector(image2)
    corr = cv2.compareHist(vect1, vect2, cv2.HISTCMP_CORREL)
    return corr #np.linalg.norm(vect1-vect2)
    
def get_style_vector(image_path):
    vgg19_style = vgg_style_layers()
    img = image.load_img(image_path, target_size=(512, 512))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    outputs = vgg19_style(x)
    style_outputs = [gram_matrix(style_output)
                         for style_output in outputs]
    style_out = tensor_to_array(style_outputs[2]).ravel()
    return style_out

def get_style_corr(image1, image2):
    vect1 = get_style_vector(image1)
    vect2 = get_style_vector(image2)
    corr = cv2.compareHist(vect1, vect2, cv2.HISTCMP_CORREL)
    return corr #np.linalg.norm(vect1-vect2)

def get_corr_mat(images1, images2, method = 'colour'):
    image_list, corr_dict = [], {}
    for im in images1:
        if im not in image_list:
            image_list.append(im)
    for im in images2:
        if im not in image_list:
            image_list.append(im)

    for im in image_list:
        if method == 'colour':
            corr_dict[im] =  get_colour_vector(im)
        elif method == 'content':
            corr_dict[im] =  get_content_vector(im)
        elif method == 'style':
            corr_dict[im] =  get_style_vector(im)

    size1 = len(images1)
    size2 = len(images2)
    corrs = np.zeros((size1,size2))
    for i in range(size1):
        for j in range(size2):
            im1 = images1[i]
            im2 = images2[j]
            corrs[i,j] = cv2.compareHist(corr_dict[im1], corr_dict[im2], cv2.HISTCMP_CORREL)
    return corrs
```

```python
context_images = ['ColourTests/0/'+val for val in os.listdir('ColourTests/0')]
method = 'colour'
reduction = 'tsne'#pca
ref_images = 10 # how many reference images to look at
components = 2
ref_stds = 0.35 # consistency of colour in reference images 0, only images with below average distance to one another are kept
start_point = np.zeros(components) # np.random.uniform(low=-40, high=40, size=(components)) # 

#print(context_images)
vects = []
for img in context_images:
    if method == 'colour':
        try:
            vect = get_colour_vector(img)
        except:
            print(img)
    vects.append(vect)

arr = np.vstack( vects )
#print(arr.shape)

if reduction == 'pca':
    pca = PCA(n_components=components)
    reduced = pca.fit_transform(arr)
elif reduction == 'tsne':
    pca = PCA(n_components=50)
    pca_red = pca.fit_transform(arr)
    pca_red = np.float64(pca_red)
    tsne = TSNE(n_components=components, random_state=0, method='exact')
    #tsne = TSNE(n_components=2, learning_rate='auto', init='random')
    reduced = tsne.fit_transform(pca_red)

dists = []
for i in range(len(context_images)):
    dists.append(np.linalg.norm(reduced[i]-start_point))
#print(dists)

reference_images = []
ordered_ims = [x for _, x in sorted(zip(dists, context_images))]
ordered_coords = [x for _, x in sorted(zip(dists, reduced))]
dists.sort()

score = sum(dists[:ref_images])/ref_images
print('consistency score at ' + str(start_point) + ': ' + str(score))

dist_mat = np.zeros([ref_images, ref_images])
for i in range(ref_images):
    for j in range(i, ref_images):
        if i != j:
            vect1 = ordered_coords[i]
            vect2 = ordered_coords[j]
            dist = np.linalg.norm(vect1-vect2)
            dist_mat[i,j] = dist
            dist_mat[j,i] = dist
            
average_dist = np.mean(dist_mat)
std_dist = np.std(dist_mat)

#print(dist_mat)
print('average dist: : ' + str(average_dist))
print('standard deviation: ' + str(std_dist))

for j in range(ref_images):
    im_ave_dist = np.sum(dist_mat[j,:])/ref_images
    remove = False
    if (im_ave_dist>average_dist+ref_stds*std_dist):
        remove = True
    
    print('image: ' + ordered_ims[j])
    print('coord: ' + str(ordered_coords[j]))
    print('distance to starting point: ' + str(dists[j]))
    print('average distance to other ref images: ' + str(im_ave_dist))
    print('excluded: ' + str(remove))
    
    plt.figure()
    img = plt.imread(ordered_ims[j])
    imgplot = plt.imshow(img)
    plt.show()
              
```
